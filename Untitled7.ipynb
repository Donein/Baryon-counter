```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
bse_data = 'RELIANCE' + '.BO'
data = yf.download(bse_data, start='2014-01-01', end='2024-04-30')
data.reset_index(inplace=True)
```

```python
data.tail()
```

```python
data.drop('Adj Close', axis=1, inplace=True)
data.head()
```

```python
data['Close'][0]
```

```python
len(data)
```

```python
len(data['Close'])
```

```python
data['Movement'] = 0
l = len(data['Close'])
for i in range(1, l):
  if data['Close'][i-1] < data['Close'][i]:
    data['Movement'][i-1] = 1
  else:
    data['Movement'][i-1] = 0

data.head()
```

```python
data.tail()
```

```python
data['Movement'].value_counts()
```

```python
data['Strength'] = 0

count_1 = 0
count_0 = 0

# Iterate through column A and calculate values for column B
for i, value in enumerate(data['Movement']):
    if value == 1:
        count_1 += 1
        data.at[i, 'Strength'] = count_1
        if i > 0 and data['Movement'][i - 1] == 0:
            count_0 = 0
    else:
        count_0 += 1
        data.at[i, 'Strength'] = -count_0
        if i > 0 and data['Movement'][i - 1] == 1:
            count_1 = 0


data.head()
```

```python
data.tail()
```

```python
data['Action'] = 0
for i in range(l-1):
  if data['Strength'][i] > 0 and data['Strength'][i+1] > 0:
    data['Action'][i] = 1
  elif data['Strength'][i] < 0 and data['Strength'][i+1] < 0:
    data['Action'][i] = -1
  else:
    data['Action'][i] = 0

data.head()
```

```python
df = data.drop(['Movement','Strength'], axis=1)
df.head()
```

```python
df.drop('Date', axis=1, inplace=True)
df.head()
```

```python
X = df.drop('Action', axis=1)
Y = df['Action']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV, GridSearchCV

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
sns.pairplot(df, hue='Action')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = KNeighborsClassifier(n_neighbors=2).fit(X_train, Y_train)
model1.score(X_test, Y_test)
```

```python
model2 = DecisionTreeClassifier().fit(X_train, Y_train)
model2.score(X_test, Y_test)
```

```python
model3 = RandomForestClassifier(n_estimators=1000)
model3.fit(X_train, Y_train)
```

```python
model3.score(X_test, Y_test)
```

```python
neighbors = range(1, 11)
scores = []
for i in neighbors:
  model1.set_params(n_neighbors=i)
  model1.fit(X_train, Y_train)
  scores.append(model1.score(X_test, Y_test))

scores
```

```python
models = {'KNN': KNeighborsClassifier(),
          'Dec_Tree': DecisionTreeClassifier(),
          'Ran_For': RandomForestClassifier()}

model_scores = {}
def model_score(models, X_train, X_test, Y_train, Y_test):
  for model_name, model in models.items():
    model.fit(X_train, Y_train)
    model_scores[model_name] = model.score(X_test, Y_test)

  return model_scores


```

```python
knn_grid = {'n_neighbors':[2],
            'weights':['uniform', 'distance'],
            'algorithm':['ball_tree', 'kd_tree', 'brute'],
            'p':[1]}

```

```python
knn_rs = RandomizedSearchCV(KNeighborsClassifier(),
                            param_distributions=knn_grid,
                            cv=5,
                            n_iter=200,
                            verbose=True)

knn_rs.fit(X_train, Y_train)

# knn_best = knn_rs.best_estimator_
knn_rs.score(X_test, Y_test)
```
